```python
categories = {
    'Design&Graphics': ['Designer', 'Artist', '3D', '2D', 'Animator', 'animation', 'Visual', 'VFX', 'Graphic', 'UI/UX', 'UI', 'UX', 'Level Designer', 'Character Artist', 'Environment Artist'],
    'QA': ['QA', 'tester', 'testing', 'Quality Assurance', 'Quality Control'],
    'Analytics': ['Analytics', 'analyst', 'Data Scientist', 'Data Engineer'],
    'HR': ['HR', 'Talent', 'Recruiter', 'Human resources', 'People', 'HR Manager', 'HR Specialist'],
    'Director': ['Director', 'Executive Producer', 'Chief Technology Officer', 'CTO', 'Studio Head', 'Head', 'Lead'],
    'Developers': ['Programmer', 'Developer', 'Software Engineer', 'Gameplay Programmer', 'Game Developer', 'Scripter', 'C++ Developer'],
    'Creative': ['Audio', 'Producer', 'Media', 'Creative', 'Writer', 'Sound', 'Narrative Designer', 'Game Writer','Copywriter'],
    'Marketing': ['Manager', 'Product', 'Marketing', 'Marketer', 'User', 'owner', 'Digital', 'Community Manager', 'Social Media Manager', 'Brand Manager'],
    'Game Design': ['Game Designer', 'Level Designer', 'System Designer', 'Combat Designer', 'Narrative Designer', 'Gameplay Designer'],
    'Project Management': ['Project Manager', 'Producer', 'Scrum Master', 'Project Coordinator', 'Agile Coach'],
    'Machine learning': ['AI', 'machine learning', 'VR', 'Artificial Intelligence', 'Virtual Reality', 'Augmented Reality']
}
```

```python
# Function to categorize based on Title
def categorize_title(title):
    for category, keywords in categories.items():
        for keyword in keywords:
            if keyword.lower() in title.lower():
                return category
    return 'Other'
```

```python
def parse_salary(salary):
    if pd.isna(salary):
        return (None, None)
    salary = salary.replace('\xa0', ' ').replace('$', '').replace(',', '')
    if 'від' in salary:
        min_salary = int(salary.split('від')[1].strip())
        return (min_salary, None)
    elif 'до' in salary:
        max_salary = int(salary.split('до')[1].strip())
        return (None, max_salary)
    elif '–' in salary:
        min_salary, max_salary = map(int, salary.split('–'))
        return (min_salary, max_salary)
    else:
        value = int(salary.strip())
        return (value, value)
```

```python
dou['Salary Min'], dou['Salary Max'] = zip(*dou['Salary'].apply(parse_salary))
```

```python
months_ua = {
    'січня': '01', 'лютого': '02', 'березня': '03', 'квітня': '04', 'травня': '05',
    'червня': '06', 'липня': '07', 'серпня': '08', 'вересня': '09', 'жовтня': '10',
    'листопада': '11', 'грудня': '12'
}
```

```python
def parse_ukrainian_date(date_str):
    for ua_month, month_num in months_ua.items():
        if ua_month in date_str:
            date_str = date_str.replace(ua_month, month_num)
            break
    return pd.to_datetime(date_str, format='%d %m %Y')
```

```python
dou['Date'] = dou['Date'].apply(parse_ukrainian_date)

dou['Published'] = dou['Date'].dt.month
```

```python
def parse_salary_djinni(salary):
    if pd.isna(salary):
        return (None, None)

    # Remove non-breaking spaces and currency symbols
    salary = salary.replace('\xa0', ' ').replace('$', '').replace(',', '').strip()

    # Handling 'from'
    if 'from' in salary:
        min_salary = int(re.search(r'\d+', salary.split('from')[1]).group())
        return (min_salary, None)

    # Handling 'to'
    elif 'to' in salary:
        max_salary = int(re.search(r'\d+', salary.split('to')[1]).group())
        return (None, max_salary)

    # Handling range with '-'
    elif '-' in salary:
        min_salary, max_salary = map(str.strip, salary.split('-'))
        return (int(min_salary), int(max_salary))

    # Handling exact value
    else:
        value = int(salary.strip())
        return (value, value)
```

Correlation map

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df_label = djinni.copy()
text_columns = ['Category', 'Occupation', 'Experience', 'Eng level', 'Company', 'Company type']

for column in text_columns:
    df_label[column + '_encoded'] = label_encoder.fit_transform(df_label[column])

# Select numeric columns and encoded text columns
numeric_columns = ['Salary Min', 'Salary Max', 'Views', 'Applied']
encoded_columns = [column + '_encoded' for column in text_columns]
all_columns = numeric_columns + encoded_columns

# Calculate correlation matrix
correlation_matrix = df_label[all_columns].corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

Salary statistics

```python
eng_order = ['Beginner/Elementary', 'Pre-Intermediate', 'Intermediate', 'Upper-Intermediate', 'Advanced/Fluent']

# Group by 'Eng level' and calculate min and mean for 'Salary Min'
salary_stats = djinni.groupby('Eng level')['Salary Min'].agg(['min', 'mean']).reset_index()

# Sort the DataFrame by the logical order
salary_stats['Eng level'] = pd.Categorical(salary_stats['Eng level'], categories=eng_order, ordered=True)
salary_stats = salary_stats.sort_values('Eng level')

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Eng level', y='value', hue='variable',
            data=pd.melt(salary_stats, id_vars='Eng level', value_vars=['min', 'mean']))
plt.title('Minimum and Mean Salary Min by English Level')
plt.xlabel('English Level')
plt.ylabel('Salary Min')
```

```python
exp_stats = djinni.groupby('Experience')['Salary Min'].agg(['min', 'mean']).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Experience', y='value', hue='variable',
            data=pd.melt(exp_stats, id_vars='Experience', value_vars=['min', 'mean']))
plt.title('Minimum and Mean Salary Min by Experience')
plt.xlabel('Experience')
plt.ylabel('Salary Min')
plt.show()
```

*   On Djinni, the median number of applications per vacancy is 12, with an average of 85 views per vacancy. On average, 28 people apply to each job posting.
*   The most popular work locations are remote (over 430 on Dou), Kyiv (around 300 on Dou and 153 on Djinni), and abroad (around 100 on Dou and 215 in Poland on Djinni).
*   Positions requiring advanced English, especially on Djinni, offer significantly higher minimum salaries compared to those requiring lower levels of English proficiency.
